```python
from transformers import pipeline

model_name = "doniNlp/AcehXBERT-sentiment"

AcehXBERT_classifier = pipeline("text-classification",model=model_name)

label_map = {
    "LABEL_0": "Negative",
    "LABEL_1": "Neutral",
    "LABEL_2": "Positive"
}

result = AcehXBERT_classifier("Rugo jak sikula manyang menyo utak kah lambat")[0]
result["label"] = label_map[result["label"]]

result
```
